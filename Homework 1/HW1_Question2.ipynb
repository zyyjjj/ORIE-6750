```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Method 1
N=5
p=1/3
theta=np.random.binomial(1,p)
q = 0.75 if theta ==1 else 0.5

ys = np.zeros(N)
for i in range(N):
    ys[i]=np.random.binomial(1,q)
```

```python
# do it 10,000 times
N, p=5, 1/3
Nsim = 10000
ys=np.zeros(Nsim)
for n in range(Nsim):
    theta=np.random.binomial(1,p)
    q = 0.75 if theta ==1 else 0.5
    ys[n]=np.random.binomial(N,q)
    
bins = np.arange(0, 6.5) - 0.5
plt.hist(ys,bins, rwidth=0.8)
plt.title('Distribution of $y_1+\cdots+y_5$ over $10^4$ trials (Method 1)')
plt.xlabel('Value of $y_1+\cdots+y_5$')
plt.ylabel('Counts')
```
